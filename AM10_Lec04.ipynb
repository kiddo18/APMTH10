```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

# Taylor expansion/series to approximate physical movement over time

```python
T = 10
Nt = 100
t = np.linspace(0,T,Nt)

m = 5.0
govm=10/m # g over m
gamovm=1.0/m # gamma over m

x = np.zeros(Nt)
v = np.zeros(Nt)
a = np.zeros(Nt) # acceleration

x[0] = 50.0
v[0] = 0.0

dt = t[1] - t[0]
dt2 = dt**2

for n in range(Nt - 1):
    a[n] = -govm + 0.5 * gamovm * v[n] ** 2
    x[n+1] = x[n]+ v[n] * dt + 0.5 * a[n] * dt2
    v[n+1] = (x[n+1] - x[n])/dt
    
plt.plot(t,x,"b") # pos vs time
plt.show()
plt.plot(t,v, "r") # velo vs time
plt.show()
plt.plot(t,a, "g") # acceleration vs time
plt.show()
```

# Derivative 

```python
def fofx(x,n):
    return x**n

def d1fofx(x,n):
    return n*(x**(n-1))
```

```python
# Init
Nx = 10
xmin =0.0
xmax = 10.0
x = np.linspace(xmin, xmax, Nx)
dx = x[1]-x[0]

# Assign
fx = fofx(x,2)
d1fx = d1fofx(x,2)

d1fnum = np.zeros(Nx)

for n in range(1,Nx-1):
    d1fnum[n] = (fx[n+1]-fx[n-1])/(2.0*dx)
    
d1fnum[0] = (fx[1] - fx[0])/dx # leftmost index
d1fnum[Nx-1] = (fx[Nx-1]-fx[Nx-2])/dx # rightmost index
```

```python
plt.plot(x,d1fnum,'ro') # approximation
plt.plot(x,d1fx,'b-') # actual

# at the end point, the approximation relies on 1 side (either left of right) and not the average
```
