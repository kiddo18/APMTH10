## 1 - Printing in Python

There are basically two ways of printing strings and variables in Python. 
- **With the percentage operator (%)**. In the example below, `%4.2f` is a format description for a float number indicating that the total number of digits are 4 with 2 decimal points. The `f` indicates a float and the format is 
```
%{width}.{precision}
```
- **With the format (f) string:** which is a more intuitive way of including variables in printing strings. With the f-strings we need not use the % operator.

````python
# what you already know
name = "Bob"
age = 24.4567
hobby = 'painting'
print('Hello, I am %s, I am %4.2f years old, and I like %s.' % (name, age, hobby))
````

````python
# format strings
name = "Bob"
salary = 240000.45675
print(f'Hello, I am {name}, I earn ${salary:8.3f} a year, and I like {hobby}.')
````

When asked to cut off decimal points, formatting rounds the number. We can perform rounding ourselves by using `np.round`,

````python
np.round(240000.45675, 3) # round to 3 decimal points
````

>**EXERCISE 1**: Let's write the Python functions for the logistic and hyperbolic math functions. Use `np.exp()` and `np.tanh()`. Choose the x points and the necessary parameters a, b, c as floats. 
* Logistic function:
  \begin{align*}
    f\left(x\right) = \dfrac{1}{1 + be^{-ax}}
  \end{align*}
  where $a$ and $b$ are parameters.<BR><BR>
* Hyperbolic tangent:
  \begin{align*}
    g\left(x\right) = b\tanh\left(ax\right) + c
  \end{align*}

**Call** the functions and print the array as the output.

````python
# your answer here
def logisticfunc(x, a, b):
    '''Logistic function'''
    
    return (1 + b * np.exp(- a * x))**-1
print(f"Logistic function: {logisticfunc(0,0,0)}")

def hyper(x, a, b, c):
    ''' hyperbolic function'''
    
    return (b * np.tanh(a * x) + c)
print(f"Hyperbolic function: {hyper(0,0,0,0)}")
````

### Let's start with `pyplot`

We will start with a simpler interface called `pyplot` which is introduced with the command 

````python
import matplotlib.pyplot as plt
````

`pyplot` has the notion of the 'active' axes which is a part of a figure, one at a time. You cannot refer to different plots by name. This works well with simple plots and for making quick figures. `Pyplot` works with both lists and numpy arrays.

It creates all plots in the same figure which is good for comparison purposes. 

**NOTE:** The format string in `pyplot` consists of a part for color, marker and line:

fmt = '[color] [marker] [line] '

See here for details on choices for [these parameters](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.plot.html)

####  Let's draw a simple plot 

````python
# Define array of variables x,y
time = np.array([1, 2, 3.2, 4.1, 5])
grades1 = np.array(['B-', 'B', 'B', 'B+', 'A'])
grades2 = np.array(['B-', 'B-', 'B-', 'B', 'B+'])
grades3 = np.array(['B+', 'B+', 'B+', 'B+', 'B+'])

# Make plot
fig = plt.figure()
plt.plot(time, grades1, 'go', label='eleni', linewidth=1)
# 'ro-' if you want the points connected
plt.plot(time, grades2, 'ro-', label='george', linewidth=1)
plt.plot(time, grades3, 'b-', label='john', linewidth=1)

# Customising the plot - add labels
plt.xlabel('Time in semester')
plt.ylabel('Grade')
plt.grid()
plt.title('My Grades as the Semester Goes By')
plt.legend();
````

### Beyond `pyplot`. Plotting using the Axes interface 

`pyplot` is good for fast plotting and you can also create subplots but it's a clunky way because there is only one "active" plot at a time. You need to keep track of the active plot, on which all `plt` commands will be applied.

A more flexible way is to use the object-oriented interface containing *methods* to *Figure* and *Axes* objects. Do not worry if you do not know what object oriented programming is. Think of *Figure* as the whole figure "umbrela" and *Axes* as an array of your individual figures. 

An example will make this more concrete. We will create the exact plot as before but with the *Axes* interface. I have the previous commands commented out so you can compare.

````python
# fig = plt.figure()
fig, ax = plt.subplots(1,1, figsize=(10,6))

x = np.linspace(-1.0, 1.0, 100)

# define the parameters
a = 2.
b = 1.
c = 1.
d = 50.


y1 = logisticfunc(x, a, b)
y2 = hyper(x, a, b, c)

# plt.plot(x, y1, 'g--')
ax.plot(x, y1, 'g--')

# plt.plot(z, y2, '.r')
ax.plot(x, y2, '.r')

# plt.xlabel('x')
ax.set_xlabel(r'$x$', fontsize=14) # Note the use of TeX for mathematical rendering

# plt.ylabel('y(x)')
ax.set_ylabel(r'$f(x)$', fontsize=14) 

# plt.grid(True) 
ax.grid(True, lw=1.5, ls='--', alpha=0.75)

# plt.title('Logistic Function');
ax.set_title('Logistic Function', fontsize=14);
````

#### What if we want to create multiple subplots?

You need to use this command
```Python
fig, ax = plt.subplots(num_rows, num_cols, figsize=(20,6))
```
Where `num_rows` are the rows in our subplot array and `num_cols` are the columns.

And then refer to each plot as a array element.
```Python
ax[i,j].plot(z, y, 'g--')
```
Where $i$ is the row and $j$ the column.

Let's create two plots next to each other. This is a 1D array of plots that has only 1 row. So we only need one index ($i$).

````python
# fig = plt.figure()
fig, ax = plt.subplots(1,2, figsize=(20,6))

ax[0].plot(x, y1, 'g--')
ax[1].plot(x, y2, '.r');

# Titles
ax[0].set_title('Logistic Function', fontsize=14);
ax[1].set_title("Hyperbolic tangent Function", fontsize=14)
````

### Error Bars

Sometimes we get measurements from an instrument or sensor that include some error. When plotting the function we can include a visual of this error using `matplotlib.pyplot.errorbar(x,y,yerr=,xerr=)` 

x, y define the data locations, xerr, yerr define the errorbar sizes. By default, this draws the data markers/lines as well the errorbars. Use fmt='none' to draw errorbars without any data markers. 

````python
x,y,yerr = np.loadtxt("my_data_err.csv",  delimiter = " ", unpack=True)
fig, ax = plt.subplots(1,1, figsize=(20,6))
ax.errorbar(x,y , yerr=yerr, label='both limits (default)');
````

### Histograms

Histograms are used to visualize the distribution of one variable. For example, let's say we want to see the distribution of people's heights in a polulation. Let's load the dataset and plot the histogram.

````python
s = np.array([3,3,3,4,4,2,2,2,5,5,5,6,6,1,1,1,3,3,3,3,3,6,6,6,6,3,3,3,3])
fontsize = 15
bins = 15
plt.hist(s, bins=bins, alpha=0.3, #normed=True, 
         color='steelblue',
         edgecolor='none');
plt.title(f'Just Numbers', fontsize=fontsize);
plt.xlabel('points', fontsize=fontsize) 
plt.ylabel('number of points', fontsize=fontsize)
plt.show();
````
