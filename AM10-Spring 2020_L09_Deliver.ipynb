### 1. Random numbers

How can a deterministic machine like a computer produce "random" numbers? They are not really random, they are generated by an algorithm, but they satisfy all the conditions of true random numbers within some approximation.  They are called "pseudo-random" numbers.

The most common algorithm is the modulo algorithm:

r[n+1] = mod[(A * r[n] + B),M]/M, A,B,M: large integers, and need a seed to get started.

This produces random numbers in the range [0,1).  We can then shift and scale them to cover unifromly any interval. 

These look good for a while, but eventually they start repeating, as you can easily show if you choose the values of A, B, M to be small (for larger values this problem is delayed, but it NOT avoided).

Moreover, if you go to higher dimensions, like generating pairs of numbers (x,y) to cover a region of the plane uniformly, you discover deeper problems. 

#### It takes hard work to create a good RNG!!!

```python
# Example 1: build our own RNG based on modulo

N=100
A=1000
B=400
M=197
seed =21

rnum = np.zeros(N)
rnum[0] = seed

for n in range(0, N-1):
    rnum[n+1] = np.mod(A * rnum[n] + B, M)
    
rnum = rnum / M # bring them back to within the inverval [0,1]

plt.plot(rnum,'ro')
plt.xlabel('Iteration')
plt.ylabel('Random number in [0,1] range')
plt.show()

plt.hist(rnum)
plt.show()
```

### These do not look bad at all!

```python
# Example 2: use numbers from our RNG to cover the plane as (x,y) coordinates

N=100
A=1000
B=400
M=895
seed1=5
seed2=59

xnum = np.zeros(N)
ynum = np.zeros(N)
xnum[0] = seed1
ynum[0] = seed2

for n in range(0, N-1):
    xnum[n+1] = np.mod(A * xnum[n] + B, M)
    ynum[n+1] = np.mod(A * ynum[n] + B, M)
    
xnum = xnum / M
ynum = ynum / M

plt.plot(xnum, ynum,'ro')
plt.xlabel('x-random number in [0,1] range')
plt.ylabel('y-random number in [0,1] range')
plt.show()
```

### These are obviously a disaster!

```python
# Example 4: calculate pi!
# REWORK THIS BY STARTING WITH THE INNER LOOP

pisum = 0.
N = 1000
x = np.random.rand(N)
y = np.random.rand(N)

for n in range(N):
    d = np.sqrt(x[n]**2 + y[n]**2)
    if (d < 1.):
        pisum = pisum + 1.
        
result = (4. * pisum) / N # ratio of number of points inside the circle / total number of points in the square
print(result)
```

```python
# Example 5: calculate integral of x^3 from x=0 to 1. MONTE CARLO INTERGRATION
x3sum = 0.
N = 1000
x = np.random.rand(N)
y = np.random.rand(N)

for n in range(N):
    y3 = x[n] ** 3 # The curve
    if (y[n] < y3): # random number much less than calculated number to calculate area under the curve
        x3sum += 1.
        
result = x3sum / N # ratio of number of points inside the circle / total number of points in the square
print(result)
```
