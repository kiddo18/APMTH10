### 1. First-order Differential Equation: Eurler's method

```python
xmin=.5
xmax=2.5

Nx=100
x=np.linspace(xmin,xmax,Nx)
dx=x[1]-x[0]

u=np.zeros(Nx)
f=np.zeros(Nx)

utrue=x*np.exp(-x)

u[0] = xmin*np.exp(-xmin) # We need to know/be given the first value 

for n in range(0, Nx-1):
    f[n] = ((1./x[n])-1.) * u[n] #(f is RHS of DE and the rest is on the LHS)
    u[n+1] = u[n] + dx * f[n] # Euler's method
    
plt.plot(x,u,'bo')
plt.plot(x, utrue,'r-')
```

### 2. Eigenfunction

```python
# Apply Euler's method to the eigenfunction problem

Nx = 50
xmin =0.
xmax=3.
x=np.linspace(xmin,xmax,Nx)
dx = x[1] - x[0]
u1 = np.zeros(Nx)
f=np.zeros(Nx)

a=2.
Afact=1. # coefficient of function u(x) = A*exp(x) 
u1[0]=Afact#--> Afact == u1[0] because exp(0) == 1
utrue=Afact*np.exp(a*x) # actual solution to DE

for n in range(0,Nx-1):
    f[n] = a* u1[n] # du/dx on LHS; a*u
    u1[n+1] = u1[n] + dx * f[n] # Euler's method

plt.plot(x,u1,'bo')
plt.plot(x, utrue,'r-')
```

### 2a. Modified Euler's method

```python
Nx = 50
xmin =0.
xmax=3.
x=np.linspace(xmin,xmax,Nx)
dx = x[1] - x[0]
u2 = np.zeros(Nx)
f=np.zeros(Nx)

a=2.
Afact=1. # coefficient of function u(x) = A*exp(x) 
u2[0]=Afact#--> Afact == u1[0] because exp(0) == 1
utrue=Afact*np.exp(a*x) # actual solution to DE

for n in range(0,Nx-1):
    f[n] = a * u1[n] # du/dx on LHS; a*u
    f[n+1] = a * (u2[n] + dx * f[n])
    u2[n+1] = u2[n] + 0.5 * dx * (f[n] + f[n]) # Euler's modified method

plt.plot(x,u1,'go')
plt.plot(x,u2,'bo')
plt.plot(x, utrue,'r-')
plt.show()

# difference percent
plt.plot(x,100*(u1-utrue)/utrue, 'yo', label='% diff Euler original')
plt.plot(x,100*(u2-utrue)/utrue, 'bo', label='% diff Euler modified')
plt.legend(loc='lower left', frameon=False)
plt.show()
```
